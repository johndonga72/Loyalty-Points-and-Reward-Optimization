calculate the key points of deposit

```python
dep_grouped = deposit.groupby(['User Id', 'Datetime','Time','slots']).agg(
    deposit_amount=('Amount', 'sum'),
    deposit_count=('Amount', 'count')
).reset_index()
```

```python
 dep_grouped
```

calculate the key points of withdraw

```python
with_grouped = deposit.groupby(['User Id', 'Datetime','Time','slots']).agg(
    withdraw_amount=('Amount', 'sum'),
    withdraw_count=('Amount', 'count')
).reset_index()
```

```python
with_grouped
```

```python

game_grouped = user.groupby(['User ID', 'Datetime', 'Time','slots']).agg(
    games_played=('Games Played', 'sum')
).reset_index()
```

```python
game_grouped
```

Rename all user_id columns to lowercase and strip whitespaces


```python
dep_grouped.rename(columns=lambda x: x.strip().lower(), inplace=True)
with_grouped.rename(columns=lambda x: x.strip().lower(), inplace=True)
game_grouped.rename(columns=lambda x: x.strip().lower(), inplace=True)
```

TO MERGE THE ALL THE TABLES INTO ONE TABLE FOR CALCULATING THE LOYAL POINTS 

```python
full_table = pd.merge(dep_grouped, with_grouped, on=['user id', 'datetime', 'slots','time'], how='outer')
full_table = pd.merge(full_table, game_grouped, on=['user id', 'datetime', 'slots','time'], how='outer')
```

```python
full_table
```

```python
full_table.head()
```

```python
full_table = full_table.fillna(0)
```

Calculate Loyalty Points

```python

full_table['loyalty_points'] = (
    0.01 * full_table['deposit_amount'] +
    0.005 * full_table['withdraw_amount'] +
    0.001 * full_table[['deposit_count', 'withdraw_count']].apply(lambda x: max(x['deposit_count'] - x['withdraw_count'], 0), axis=1) +
    0.2 * full_table['games_played']
)
```

After calculating the loyalty points for the whole month find out which 50 players are at the top of the leaderboard. The company has allocated a pool of Rs 50000 to be given away as bonus money to the loyal players.


```python
# Sort within each month
full_table1 = full_table.sort_values(by=['month_name', 'loyalty_points', 'games_played'], ascending=[True, False, False])
full_table1['rank'] = full_table1.groupby('month_name').cumcount() + 1
```

```python
full_table1.head()
```

Suggest a suitable way to divide the allocated money keeping in mind the following points
-- here i follow loyality points because loyality points calcualted by deposit, number of playing games and with draw, so my opinion is to take loyality points. so  i consider loyality points

take  top 50 each month

```python
top50_per_month = full_table1[full_table1['rank'] <= 50].copy()
```

```python
top50_per_month.head()
```

distribute the bonus based on the loyality points

```python
bonus_pool = 50000
top50_per_month['bonus_money'] = top50_per_month.groupby('month_name')['loyalty_points'].transform(
    lambda x: (x / x.sum()) * bonus_pool
).round(2)
```

```python
top50_per_month.head()
```

```python
final_output = top50_per_month[['month_name', 'user id', 'loyalty_points', 'games_played', 'rank', 'bonus_money']]
```
